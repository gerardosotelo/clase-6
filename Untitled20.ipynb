#Problema 1:

```python
class Playlist:
    # Representa una lista de canciones (implementada con lista de Python)
    def __init__(self):
        self.songs = []  # lista que guarda las canciones

    def append(self, x):
        # Agrega una canción al final
        self.songs.append(x)

    def insert(self, i, x):
        # Inserta canción en la posición i
        self.songs.insert(i, x)

    def remove(self, i):
        # Elimina canción en la posición i
        if 0 <= i < len(self.songs):
            self.songs.pop(i)

    def get(self, i):
        # Obtiene la canción en la posición i
        if 0 <= i < len(self.songs):
            return self.songs[i]
        return None

    def set(self, i, x):
        # Cambia la canción en la posición i
        if 0 <= i < len(self.songs):
            self.songs[i] = x

    def size(self):
        # Devuelve el tamaño de la playlist
        return len(self.songs)

    def search(self, substring):
        # Busca canciones que contengan cierto texto
        return [s for s in self.songs if substring.lower() in s.lower()]

    def move(self, i, j):
        # Mueve canción de la posición i a la posición j
        if 0 <= i < len(self.songs):
            song = self.songs.pop(i)
            self.songs.insert(j, song)

```

```python
def playlist_demo():
    pl = Playlist()
    while True:
        print("\n=== PLAYLIST ===")
        print("1. Agregar canción")
        print("2. Insertar en posición")
        print("3. Eliminar canción")
        print("4. Mostrar playlist")
        print("5. Buscar canción")
        print("6. Mover canción")
        print("0. Salir")
        op = input("Elige una opción: ")
        if op == "1":
            song = input("Nombre de la canción: ")
            pl.append(song)
        elif op == "2":
            i = int(input("Posición: "))
            song = input("Nombre de la canción: ")
            pl.insert(i, song)
        elif op == "3":
            i = int(input("Posición: "))
            pl.remove(i)
        elif op == "4":
            print("Playlist:", [pl.get(i) for i in range(pl.size())])
        elif op == "5":
            s = input("Texto a buscar: ")
            print("Resultados:", pl.search(s))
        elif op == "6":
            i = int(input("Posición actual: "))
            j = int(input("Nueva posición: "))
            pl.move(i, j)
        elif op == "0":
            break
```

```python
playlist_demo()
```

```python
class Stack:
    # Implementación básica de una pila (LIFO: Last In, First Out)
    def __init__(self):
        self.data = []  # lista que guarda los elementos

    def push(self, x):
        # Inserta un elemento en la pila
        self.data.append(x)

    def pop(self):
        # Saca el último elemento insertado
        return self.data.pop() if not self.is_empty() else None

    def peek(self):
        # Devuelve el último elemento sin sacarlo
        return self.data[-1] if not self.is_empty() else None

    def is_empty(self):
        # Revisa si la pila está vacía
        return len(self.data) == 0

    def size(self):
        # Devuelve el número de elementos en la pila
        return len(self.data)


class TextEditor:
    # Editor de texto con soporte de "deshacer" (undo) y "rehacer" (redo)
    def __init__(self):
        self.text = ""              # texto actual
        self.undo_stack = Stack()   # pila para operaciones undo
        self.redo_stack = Stack()   # pila para operaciones redo

    def type(self, txt):
        # Agregar texto al final
        self.undo_stack.push(("delete", len(txt)))  # guardar acción inversa
        self.text += txt
        self.redo_stack = Stack()  # limpiar pila de redo

    def delete(self, k):
        # Borrar últimos k caracteres
        deleted = self.text[-k:]
        self.undo_stack.push(("type", deleted))  # guardar acción inversa
        self.text = self.text[:-k]
        self.redo_stack = Stack()

    def undo(self):
        # Deshacer la última acción
        if not self.undo_stack.is_empty():
            action, value = self.undo_stack.pop()
            if action == "delete":
                deleted = self.text[-value:]
                self.redo_stack.push(("type", deleted))
                self.text = self.text[:-value]
            elif action == "type":
                self.redo_stack.push(("delete", len(value)))
                self.text += value

    def redo(self):
        # Rehacer una acción deshecha
        if not self.redo_stack.is_empty():
            action, value = self.redo_stack.pop()
            if action == "delete":
                deleted = self.text[-value:]
                self.undo_stack.push(("type", deleted))
                self.text = self.text[:-value]
            elif action == "type":
                self.undo_stack.push(("delete", len(value)))
                self.text += value

    def content(self):
        # Devuelve el texto actual
        return self.text

```

```python
def editor_demo():
    ed = TextEditor()
    while True:
        print("\n=== EDITOR ===")
        print("Texto actual:", ed.content())
        print("1. Escribir texto")
        print("2. Borrar k caracteres")
        print("3. Undo(desacer)")
        print("4. Redo(reacer)")
        print("0. Volver al menú principal")
        op = input("Elige una opción: ")

        if op == "1":
            txt = input("Texto a escribir: ")
            ed.type(txt)
        elif op == "2":
            k = int(input("Cantidad de caracteres a borrar: "))
            ed.delete(k)
        elif op == "3":
            ed.undo()
        elif op == "4":
            ed.redo()
        elif op == "0":
            break
```

```python
class CircularQueue:
    # Cola circular para manejar tickets de soporte
    def __init__(self, capacity=5):
        self.data = [None] * capacity
        self.front_index = 0   # índice del primer elemento
        self.rear_index = 0    # índice donde entra el siguiente
        self.size_val = 0      # número de elementos
        self.capacity = capacity

    def is_empty(self):
        # Revisa si la cola está vacía
        return self.size_val == 0

    def size(self):
        # Devuelve la cantidad de elementos
        return self.size_val

    def enqueue(self, x):
        # Agrega un nuevo elemento al final
        if self.size_val == self.capacity:
            self._resize()  # si está llena, se redimensiona
        self.data[self.rear_index] = x
        self.rear_index = (self.rear_index + 1) % self.capacity
        self.size_val += 1

    def dequeue(self):
        # Saca el primer elemento en entrar
        if self.is_empty():
            return None
        item = self.data[self.front_index]
        self.data[self.front_index] = None
        self.front_index = (self.front_index + 1) % self.capacity
        self.size_val -= 1
        return item

    def front(self):
        # Devuelve el primer elemento sin sacarlo
        if self.is_empty():
            return None
        return self.data[self.front_index]

    def _resize(self):
        # Duplica la capacidad cuando la cola está llena
        old = self.data
        self.capacity *= 2
        self.data = [None] * self.capacity
        idx = 0
        for i in range(self.size_val):
            self.data[i] = old[(self.front_index + i) % len(old)]
            idx += 1
        self.front_index = 0
        self.rear_index = idx


class Ticket:
    # Representa un ticket de soporte
    def __init__(self, id, user, arrival, duration):
        self.id = id
        self.user = user
        self.arrival = arrival    # minuto de llegada
        self.duration = duration  # tiempo de atención


class SupportCenter:
    # Simulador del centro de soporte
    def __init__(self):
        self.queue = CircularQueue()

    def add_ticket(self, ticket):
        # Agregar ticket a la cola
        self.queue.enqueue(ticket)

    def simulate(self):
        # Simula la atención de todos los tickets
        time = 0
        order = []
        total_wait = 0
        count = self.queue.size()

        while not self.queue.is_empty():
            t = self.queue.dequeue()
            if time < t.arrival:
                time = t.arrival  # si aún no llega el ticket, avanzar el reloj
            wait = time - t.arrival
            total_wait += wait
            order.append((t.id, t.user, wait))
            time += t.duration

        avg_wait = total_wait / count if count > 0 else 0
        return order, avg_wait
```

```python
def soporte_demo():
    sc = SupportCenter()
    while True:
        print("\n=== SOPORTE TÉCNICO ===")
        print("1. Agregar ticket")
        print("2. Simular atención")
        print("0. Volver al menú principal")
        op = input("Elige una opción: ")

        if op == "1":
            id = input("ID del ticket: ")
            user = input("Usuario: ")
            arrival = float(input("Minuto de llegada: "))
            duration = float(input("Duración atención: "))
            sc.add_ticket(Ticket(id, user, arrival, duration))
        elif op == "2":
            orden, prom = sc.simulate()
            print("Orden de atención:", orden)
            print("Tiempo de espera promedio:", prom)
        elif op == "0":
            break


# ======== MENÚ PRINCIPAL ========

def main():
    while True:
        print("\n=== MENÚ PRINCIPAL ===")
        print("1. Playlist")
        print("2. Editor de Texto")
        print("3. Soporte Técnico")
        print("0. Salir")
        op = input("Elige una opción: ")

        if op == "1":
            playlist_demo()
        elif op == "2":
            editor_demo()
        elif op == "3":
            soporte_demo()
        elif op == "0":
            print("¡Adiós!")
            break

# Ejecutar
main()
```
